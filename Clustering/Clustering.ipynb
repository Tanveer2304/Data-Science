# K Means Cluster

```python
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
df.head()
```

```python
df2 = df.iloc[:,6:8].values
```

```python
#Preprocessing the data using min max scaler

#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
```

```python
#scaler.fit(df2)
#df2 = scaler.fit_transform(df2)
```

```python
df2
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++', random_state= 42)
    kmeans.fit(df2)
    
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()


```

```python
kmeans = KMeans(n_clusters=4,init='k-means++', random_state= 42)

y2 =kmeans.fit_predict(df2)
print(y2)
```

```python
np.unique(y2)
```

```python
plt.scatter(df2[y2 == 0,0], df2[y2 == 0,1], c = 'green')
plt.scatter(df2[y2 == 1,0], df2[y2 == 1,1], c = 'red')
plt.scatter(df2[y2 == 2,0], df2[y2 == 2,1], c = 'yellow')
plt.scatter(df2[y2 == 3,0], df2[y2 == 3,1], c = 'blue')


# plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='cyan', label='Centroids')
plt.title('Earning miles with non flight transactions')
plt.xlabel('non flight Bonus Miles')
plt.ylabel('non flight Bonus Transactions')
plt.show()
```

```python
#Evaluation and Performance Metrics

from sklearn.metrics import silhouette_score
```

```python
silhouette_score_coeff = []

# Iterate over different numbers of clusters
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df2)
    score = silhouette_score(df2, kmeans.labels_)
    silhouette_score_coeff.append(score)
```

```python
silhouette_score(df2, y2)
```

```python
silhouette_score_coeff
```

```python
plt.plot(range(2,11),silhouette_score_coeff)
```
